```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate a sample dataset with location information
data = {
    'Latitude': [37.7749, 37.7749, 37.7749, 34.0522, 34.0522, 34.0522, 40.7128, 40.7128, 40.7128],
    'Longitude': [-122.4194, -122.4194, -122.4194, -118.2437, -118.2437, -118.2437, -74.0060, -74.0060, -74.0060],
}

df = pd.DataFrame(data)

# Visualize the data
plt.scatter(df['Longitude'], df['Latitude'])
plt.title('Location Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Longitude', 'Latitude']])

# Visualize the clustered data
plt.scatter(df['Longitude'], df['Latitude'], c=df['Cluster'], cmap='viridis')
plt.title('Clustered Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Display the clustered data
df
```
